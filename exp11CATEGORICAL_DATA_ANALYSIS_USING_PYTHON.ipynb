```python
import pandas as pd
data={
    'order_ID':['01','02','03','04','05','06','07','08','09','010'],
    'category':['electronics','grocery','books','electronics','grocery','books','electronics','grocery','books','grocery'],
    'payment_method':['cod','upi','card','cod','card','cod','upi','upi','upi','cod'],
    'delivery type':['regular','express','express','regular','express','regular','express','regular','express','regular'],
    'customer type':['new','returning','new','returning','new','returning','new','returning','new','returning']
}
df=pd.DataFrame(data)
print(df)
```

```python
#1. frequency count
#count how many oders in each category.
category_count=df['category'].value_counts()
print(category_count)
```

```python
#unique values
unique_values=df['category'].unique()
print(unique_values)
```

```python
#3. number of unique values
num_unique_values=df['category'].nunique()
print(num_unique_values)
```

```python
#4.frequency of payment methods
payment_count=df['payment_method'].value_counts()
print(payment_count)
```

```python
#5.cross tabulation
crosstab_result=pd.crosstab(df['category'],df['payment_method'])
print(crosstab_result)
```

```python
#6.cross tabulation (delivery vs customer time)
crosstab_result=pd.crosstab(df['delivery type'],df['customer type'])
print(crosstab_result)
```

```python
#7.percentage distribution
percentage_distribution=(df['category'].value_counts(normalize=True))*100
print(percentage_distribution)
```

```python
#8A.filtering data
#example: orders from electronics category
filtered_data=df[df['category']=='electronics']
print(filtered_data)
```

```python
#8B.filtering data
#example:orders from grocery categrory
filtered_data=df[df['category']=='grocery']
print(filtered_data)
```

```python
#9 grouping
#orders by categroy and payment method
grouped_data=df.groupby(['category','payment_method']).size().reset_index(name='count')
print(grouped_data)
```

```python
#10. Sorting data
sorted_data=df.sort_values(by='order_ID')
print(sorted_data)
```

```python
#Expt11 new CSV file
#Second Dataset
df = pd.read_csv('/content/Expt11.csv')
print(df)
```

```python
#1. count students in each grade
#find how many students obtained each grade.
grade_count=df['Grade'].value_counts()
print(grade_count)
```

```python
#2.percentage distribution of grades
percentage_distribution=(df['Grade'].value_counts(normalize=True))*100
print(percentage_distribution)
```

```python
#3. number oif students in each department
department_count=df['Department'].value_counts()
print(department_count)
```

```python
#4.gender distribution
gender_count=df['Gender'].value_counts()
print(gender_count)
```

```python
#5.gender vs grade analysis
crosstab_result=pd.crosstab(df['Gender'],df['Grade'])
print(crosstab_result)
```

```python
#6.department vs gender
crosstab_result=pd.crosstab(df['Department'],df['Gender'])
print(crosstab_result)
```

```python
#7. department vs grade percentage
crosstab_result=pd.crosstab(df['Department'],df['Grade'],normalize='columns')*100
print(crosstab_result)

```

```python
#8. grouping data(department wise)
grouped_data=df.groupby(['Department']).size().reset_index(name='count')
print(grouped_data)
```
